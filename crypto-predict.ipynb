```python
train_df = pd.read_csv('./train.csv', low_memory=False, 
                       dtype={'Asset_ID': 'int8', 'Count': 'int32', 'row_id': 'int32', 'Count': 'int32', 
                              'Open': 'float64', 'High': 'float64', 'Low': 'float64', 'Close': 'float64', 
                              'Volume': 'float64', 'VWAP': 'float64'})
```

```python
train_extra = pd.read_csv('../input/g-research-crypto-forecasting/supplemental_train.csv', low_memory=False, 
                       dtype={'Asset_ID': 'int8', 'Count': 'int32', 'row_id': 'int32', 'Count': 'int32', 
                              'Open': 'float64', 'High': 'float64', 'Low': 'float64', 'Close': 'float64', 
                              'Volume': 'float64', 'VWAP': 'float64'})
```

```python
test_df = pd.read_csv('../input/g-research-crypto-forecasting/example_test.csv', low_memory = False, 
                      dtype={'Asset_ID': 'int8', 'Count': 'int32', 'row_id': 'int32', 'Count': 'int32', 
                              'Open': 'float64', 'High': 'float64', 'Low': 'float64', 'Close': 'float64', 
                              'Volume': 'float64', 'VWAP': 'float64'})
```

```python
train_df['timestamp'] = train_df['timestamp'].astype('datetime64[s]')
train_extra['timestamp'] = train_extra['timestamp'].astype('datetime64[s]')
test_df['timestamp'] = test_df['timestamp'].astype('datetime64[s]')
```

```python
train_df = train_df.reindex(range(train_df.index[0],train_df.index[-1]+60,60),method='pad')
train_extra = train_extra.reindex(range(train_extra.index[0],train_extra.index[-1]+60,60),method='pad')
test_df = test_df.reindex(range(test_df.index[0],test_df.index[-1]+60,60),method='pad')
```

```python
train_df = train_df.reset_index(drop=True)
```

```python
train_extra = train_extra.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
```

```python
##over lap discovered
train_df = train_df[train_df['timestamp']<'2021-06-13 00:00:00']
print(train_df.timestamp.max())
print(train_extra.timestamp.min())
```

```python
print(train_df.shape)
print(train_extra.shape)
```

```python
train_df.loc[:,'timestamp_y'] = train_df['timestamp'].dt.year
train_df.loc[:,'timestamp_m'] = train_df['timestamp'].dt.month
train_df.loc[:,'timestamp_d'] =  train_df['timestamp'].dt.day
train_df.loc[:,'timestamp_h'] = train_df['timestamp'].dt.hour
train_df.loc[:,'timestamp_minute'] = train_df['timestamp'].dt.minute
```

```python
train_extra.loc[:,'timestamp_y'] = train_extra['timestamp'].dt.year
train_extra.loc[:,'timestamp_m'] = train_extra['timestamp'].dt.month
train_extra.loc[:,'timestamp_d'] =  train_extra['timestamp'].dt.day
train_extra.loc[:,'timestamp_h'] = train_extra['timestamp'].dt.hour
train_extra.loc[:,'timestamp_minute'] = train_extra['timestamp'].dt.minute
```

```python
test_df.loc[:,'timestamp_y'] = test_df['timestamp'].dt.year
test_df.loc[:,'timestamp_m'] = test_df['timestamp'].dt.month
test_df.loc[:,'timestamp_d'] =  test_df['timestamp'].dt.day
test_df.loc[:,'timestamp_h'] = test_df['timestamp'].dt.hour
test_df.loc[:,'timestamp_minute'] = test_df['timestamp'].dt.minute
```
